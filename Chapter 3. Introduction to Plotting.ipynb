# Employee Salary Dataset

```python
emp_salary = pd.read_csv("employee_salary_data.csv")
```

```python
emp_salary.head()
```

```python
# remove all white spaces from right
emp_salary["JobTitle"] = emp_salary["JobTitle"].str.rstrip()
```

```python
# select Police Officer, EMT and Corrections Officer
new_emp_salary = emp_salary.loc[(emp_salary["JobTitle"]=="Police Officer") | (emp_salary["JobTitle"]=="Corrections Officer") | (emp_salary["JobTitle"]=="EMT")]
```

```python
new_emp_salary["JobTitle"].unique()
```

# 3. Multivariate 

## 1. Scatter Plot

```python
# There is no multivariate scatter plot using pandas objects
```

```python
# using matplotlib
# create a color variable based on job title
def recode_job(job):
    if job == 'Police Officer':
        return 0
    elif job == 'Corrections Officer':
        return 1
    else:
        return 2

new_emp_salary['job_color'] = new_emp_salary['JobTitle'].apply(recode_job)

scatter_plot = plt.figure()
ax = scatter_plot.add_subplot(1,1,1)
ax.scatter(x=new_emp_salary['AnnualRate'],
           y=new_emp_salary['YearToDate'],
           c=new_emp_salary['job_color'],
           alpha=0.5)
plt.show()
```

```python
scatter = sns.lmplot(x='RegularRate', y='YearToDate', data=new_emp_salary, hue='JobTitle', fit_reg=False)
plt.show()
```

### Practise

```python
# using matplotlib
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x=new_emp_salary['OvertimeRate'],
           y=new_emp_salary['Other'],
           c=new_emp_salary['job_color'],
          alpha=0.5)
plt.show()
```

```python
# using seaborn
ax = sns.lmplot(x='AnnualRate', y='YearToDate', data=new_emp_salary, hue='JobTitle', fit_reg=False)
plt.show()
```
