# 3. Grouped and Aggregated Calculations

```python
df
```

## (a) Grouped means

### (i) Average life expectancy for every year

```python
df.groupby('year')['lifeExp'].mean()
```

### Exercise: Average population for every continent

```python
df.groupby('continent')[['pop','lifeExp']].mean()
```

### (ii) Average life expectancy and GDP per capital based on year and continent

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
```

### Exercise: Average life expectancy, population and GDP per capital based on year and country

```python
df.groupby(['year','country'])[['lifeExp','pop','gdpPercap']].mean()
```

## (b) Grouped frequency counts

```python
# Return number of unique elements in the object
df.groupby('continent')['country'].nunique()
```

```python
# Returns object containing counts of unique values
df.groupby('continent')['country'].value_counts()
```
