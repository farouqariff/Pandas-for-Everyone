# 1. Pandas Series

```python
import pandas as pd
```

```python
car = pd.read_csv('C:/Users/User/Desktop/Data Science/6. Pandas for Everyone/data/car.csv')
```

```python
car.head()
```

```python
ages = 2022 - car['Year']
```

```python
ages
```

## (a) Boolean subsetting

```python
# Calculate mean
ages.mean()
```

```python
# Subset only ages more than mean
# Want return labels and indices
ages[ages > ages.mean()]
```

```python
# Subset only ages more than mean
# Want return boolean values
ages > ages.mean()
```

## (b) Operation

```python
# Vector with Vector
ages + ages
```

```python
# Vector with Integer
ages + 100
```

```python
# Vector with Vector (Different length)
# Rest of the resulting vector will be filled with NaN

ages + pd.Series([1,100])
```

# 2. Pandas DataFrame

## (a) Boolean subsetting

```python
# Subset car ages more than average
car[(2022-car['Year']) > (2022-car['Year']).mean()]
```

## (b) Operation

```python
first_4 = car[:4]
first_4
```

- When we perform an action on a __DataFrame with a scalar__, it will try to __apply the operation on each cell of the dataframe.__ 
- In this example, numbers will be multiplied by 2, and __strings will be doubled__ 

```python
first_4*2
```
